```python
def lis2dh12_from_lsb_to_celsius(lsb: int) -> float:
  if lsb & 0x8000:
    lsb -= 0x10000

  return float((lsb / 64.0) / 4.0 + 25.0);


def bytes_to_temperature(byte_l, byte_h):
  lsb = (byte_h << 8) | byte_l
  return lis2dh12_from_lsb_to_celsius(lsb)
```
